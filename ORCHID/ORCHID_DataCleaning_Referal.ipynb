```python
# Importing the libraries
import pandas as pd
import numpy as np
import os
# Ignore all warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
cwd = os.getcwd() 
folder_path = os.path.join(cwd, 'RawData')
cleaned_folder_path = os.path.join(cwd, 'CleanedCSVFiles')
```

```python
# reading csv files and storing as a dataframe
df_referral = pd.read_csv(os.path.join(folder_path,"referrals.csv"))
```

### Referral DataFrame Cleaning:

```python
# looking at the data
df_referral.head(5)
```

```python
# checking the information of the dataframe
df_referral.info()
```

```python
# List of columns to convert data type from object to timestamp
columns_to_convert = ['time_asystole', 'time_brain_death', 'time_referred', 'time_approached','time_authorized','time_procured']

# Apply pd.to_datetime to each column
for col in columns_to_convert:
    df_referral[col] = pd.to_datetime(df_referral[col], errors='coerce')

#List of columns to convert data type from float to int
col_toConvert_int = ['Procured_Year','Age']
# Apply astype() to convert
for colint in col_toConvert_int:
    df_referral[colint] = df_referral[colint].astype('Int64')
```

```python
# checking for duplicate records 
print("Duplicate Rows:\n", df_referral.duplicated().sum())
```

```python
# Checking for the value counts of 'mechanism_of_death' column for cleanup
df_referral['Mechanism_of_Death'].value_counts()
```

```python
# Created dictionary mapping to replace similar values
replace_dict = {
    'Gun Shot Wound': 'Gunshot Wound',
    'Drug / Intoxication': 'Drug Intoxication',
    'None of the Above': 'None of the above',
    'ICH/Stroke':'Intracranial Hemmorrhage / Stroke',
    'Death from Natural Causes':'Natural Causes'
}

df_referral['Mechanism_of_Death'].replace(replace_dict, inplace=True)
```

```python
# This step shows the updated value counts for 'mechanism_of_death' column
df_referral['Mechanism_of_Death'].value_counts()
```

```python
# Checking for the value counts of 'circumstances_of_death' column for cleanup
df_referral['Circumstances_of_Death'].value_counts() 
```

```python
# Created dictionary mapping to replace similar values
replace_dict_cod = {
    'Natural Causes': 'Death from Natural Causes',
    'Accident, Non-MVA': 'Non-Motor Vehicle Accident',
    'MVA': 'Motor Vehicle Accident'   
}

df_referral['Circumstances_of_Death'].replace(replace_dict_cod, inplace=True)
```

```python
# This step shows the updated value counts for 'circumstances_of_death' column
df_referral['Circumstances_of_Death'].value_counts()
```

```python
df_referral['time_procured'].value_counts()
```

```python
# Checking bloodtype and rh column to see the count of not null values before merging both the columns
df_referral[(df_referral['ABO_BloodType'].notna() & df_referral['ABO_Rh'].notna())].shape
```

```python
# Code to merge both the columns when the values are not null, This step ignores the cases where bloodtype is null and rh is not null
df_referral.loc[df_referral['ABO_BloodType'].notna() & df_referral['ABO_Rh'].notna(), 'ABO_BloodType'] = (
    df_referral['ABO_BloodType'] + '-' + df_referral['ABO_Rh']
)
```

```python
# Checking to verify the unique values after merging the columns
df_referral['ABO_BloodType'].unique()
```

```python
# Deleting the columns 'ABO_Rh' and 'Referral_DayofWeek'. 'abo_rh' values are merged so its duplicate. 
#Dropped 'Referral_DayofWeek' column as we did not find any significance towards our objective.

df_referral.drop(columns=['ABO_Rh','Referral_DayofWeek'], inplace= True)
```

```python
df_referral['WeightKg'].max()
```

```python
df_referral.query('WeightKg > 650')
```

```python
# Converting all weight values above 650kg to null. Since normal weight of person could not exceed more than that as per our search.
df_referral['WeightKg'] = df_referral['WeightKg'].apply(lambda x: np.nan if x > 650 else x)
```

```python
# rounding the weight column values to 2 decimal places
df_referral['WeightKg'] = df_referral['WeightKg'].round(2)
```

```python
df_referral['HeightIn'].max()
```

```python
df_referral.query('HeightIn > 108')
```

```python
# Converting all height values above 108inch to null. Since height of person can not exceed more than that as per our search.
df_referral['HeightIn'] = df_referral['HeightIn'].apply(lambda x: np.nan if x > 108 else x)
```

```python
# rounding the height column values to 2 decimal places
df_referral['HeightIn'] = df_referral['HeightIn'].round(2)
```

```python
# The process of procurement between January 1, 2015 and December 31, 2021. We noticed outliers before and after the mentioned date 
# in 'procured_year' column.- 10 rows
df_referral.loc[(df_referral['Procured_Year'] < 2015) | (df_referral['Procured_Year'] > 2021)]
```

```python
# Changing the outliers to null in 'procured_year' column
df_referral.loc[(df_referral['Procured_Year'] < 2015) | (df_referral['Procured_Year'] > 2021), 
       'Procured_Year'] = np.nan
```

```python
# Procured step is before transplant so when transplant column has value True, procured has to be True. There is one patient who
# does not satify the condition

df_referral.loc[(df_referral['transplanted'] == True) & (df_referral['procured'] == False)]
```

```python
# Updating procured to True based on the condition for the above patient
df_referral.loc[(df_referral['transplanted'] == True) & (df_referral['procured'] == False), 
       'procured'] = True
```

```python
# When procured_year column has year, we assume that the procurement happened so the procured column has to be True. 
# There are 39 patients who does not satisfy that condition.
df_referral.loc[(df_referral['Procured_Year'].notna()) & (df_referral['procured'] == False)]
```

```python
# Updating procured to True when procured_year column has year value.
df_referral.loc[(df_referral['Procured_Year'].notna()) & (df_referral['procured'] == False), 
       'procured'] = True
```

```python
# The authorized step is before procured so when procured is True, authorized has to be True. There are 17 patients who does not satify the condition.
df_referral.loc[(df_referral['procured'] == True) & (df_referral['authorized'] == False)]
```

```python
# Updating authorized to True when procured is True
df_referral.loc[(df_referral['procured'] == True) & (df_referral['authorized'] == False), 'authorized'] = True
```

```python
# The approached step is before authorized so when authorized is True, approached has to be True. 
# There are 30 patients who does not satify the condition.
df_referral.loc[(df_referral['authorized'] == True) & (df_referral['approached'] == False)]
```

```python
# Updating approached to True when authorized is True
df_referral.loc[(df_referral['authorized'] == True) & (df_referral['approached'] == False), 'approached'] = True
```

```python
# When 'time_brain_death' column has timestamp, it means the patient's brain_death happened. There are 4 patients whose data doesnt 
# satify the condition
df_referral.loc[(df_referral['time_brain_death'].notna()) & (df_referral['brain_death'] == False)]
```

```python
# Updating brain_death to True when time_brain_death has timestamp - There are 4 patients who did not satisfy the condition
df_referral.loc[(df_referral['time_brain_death'].notna()) & (df_referral['brain_death'] == False), 'brain_death'] = True
```

```python
# Steps for cleaning up cause_of_death_opo,cause_of_death_unos,mechanism_of_death columns
# 3054 recd - when cause_of_death_unos is null, cause_of_death_opo is not null then replace cause_of_death_unos with cause_of_death_opo value
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull()) & (df_referral['Cause_of_Death_OPO'].notna())]
```

```python
# replacing the values
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull()) & (df_referral['Cause_of_Death_OPO'].notna()),'Cause_of_Death_UNOS'] = df_referral['Cause_of_Death_OPO']
```

```python
#22980 records - when cause_of_death_unos is Other, cause_of_death_opo is not null then replace cause_of_death_unos with cause_of_death_opo value
df_referral.loc[(df_referral['Cause_of_Death_UNOS'] == 'Other') & (df_referral['Cause_of_Death_OPO'].notna())]
```

```python
# Replacing the values
df_referral.loc[(df_referral['Cause_of_Death_UNOS'] == 'Other') & (df_referral['Cause_of_Death_OPO'].notna()),'Cause_of_Death_UNOS'] = df_referral['Cause_of_Death_OPO']
```

```python
# 215 records - when cause_of_death_unos is null, mechanism_of_death is not null and None of the above then replace cause_of_death_unos with mechanism_of_death value
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull()) & (df_referral['Mechanism_of_Death'].notna()) & (df_referral['Mechanism_of_Death'] != 'None of the above')]
```

```python
# Replacing the values
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull()) & (df_referral['Mechanism_of_Death'].notna()) & (df_referral['Mechanism_of_Death'] != 'None of the above'),'Cause_of_Death_UNOS'] = df_referral['Mechanism_of_Death'] 
```

```python
# 747 records - when cause_of_death_unos is Other, mechanism_of_death is not null and None of the above then replace cause_of_death_unos with mechanism_of_death value
df_referral.loc[(df_referral['Cause_of_Death_UNOS'] == 'Other') & (df_referral['Mechanism_of_Death'].notna()) & (df_referral['Mechanism_of_Death'] != 'None of the above')]
```

```python
# Replacing the values
df_referral.loc[(df_referral['Cause_of_Death_UNOS'] == 'Other') & (df_referral['Mechanism_of_Death'].notna()) & (df_referral['Mechanism_of_Death'] != 'None of the above'),'Cause_of_Death_UNOS'] = df_referral['Mechanism_of_Death']
```

```python
# Time of asystole (i.e. cardiac death) So when time_asystole has value and cause_of_death_unos is null, replace null with 'Cardiac'
# 15708 records
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull()) & (df_referral['time_asystole'].notna())]
```

```python
# Replacing the values
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull()) & (df_referral['time_asystole'].notna()),'Cause_of_Death_UNOS'] = 'Cardiac'
```

```python
# Verifying for the values
df_referral.loc[(df_referral['Cause_of_Death_UNOS'].isnull())]
```

```python
# dropping 'cause_of_death_opo','mechanism_of_death' columns after merging the data
df_referral.drop(columns=['Cause_of_Death_OPO','Mechanism_of_Death'], inplace= True)
```

```python
# Renamed columns to maintain consistency
df_referral.rename(columns = {'Cause_of_Death_UNOS':'cause_of_death','PatientID':'Patient_ID','HospitalID':'Hospital_ID','HeightIn':'Height_In','WeightKg':'Weight_Kg'}, inplace = True)
```
